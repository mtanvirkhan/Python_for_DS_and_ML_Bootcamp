```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(5,4))
df
```

```python
df = pd.DataFrame(np.random.randn(5,4), ['A', 'B', 'C', 'D', 'E'], ['W', 'X', 'Y', 'Z'])
print(df)
print(type(df))
```

#### Column wise data show

```python
print(df['W'])
print(type(df['W']))
print(df[['W', 'X', 'Y']])
```

#### New column create

```python
# Creating new column
df['New'] = np.random.rand(5,1)
df
```

```python
# Creating new column
df['New_1'] = df['W'] + df['X']
df
```

#### New row create

```python
new_row = np.round(np.random.randn(len(df.columns)), 6)
print(new_row)
df.loc['F'] = new_row
df
```

```python
# series data
df.loc['F']
```

```python
# series data
df.iloc[2]
```

#### Column drop

```python
df.drop('New', axis=1)
```

```python
df.drop('New', axis=1, inplace=True)
df
```

```python
df.drop('F', axis=0)
```
