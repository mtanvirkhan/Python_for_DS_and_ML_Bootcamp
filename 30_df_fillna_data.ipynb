```python
import numpy as np
import pandas as pd
```

```python
dict = {
    'A': [np.nan, np.nan, 3, 4,np.nan],
    'B': [5, np.nan, 7, 8,np.nan],
    'C': [9, 10, 11, 12,25],
    'D': [np.nan, 14, 15, 16,np.nan],
    'E': [17, np.nan, 19, 20,np.nan],
    'F': [np.nan, np.nan, 23, np.nan,np.nan]
    
}
df = pd.DataFrame(dict)
df
```

```python
df.fillna('fill data')
```
