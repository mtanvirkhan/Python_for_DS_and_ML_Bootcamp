```python
import pandas as pd
```

### Concat method and append method

```python
# Dictionaries declaration
dict_profile1 = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Department': ['HR', 'Finance', 'IT'],
    'Salary': [50000, 60000, 70000]
}

dict_profile2 = {
    'Name': ['Alice2', 'Bob2', 'Charlie2'],
    'Age': [45, 50, 65],
    'Department': ['Marketing', 'Sales', 'ED'],
    'Salary': [50000, 60000, 70000]
}

dict_profile3 = {
    'Name': ['Alice3', 'Bob3', 'Charlie3'],
    'Age': [45, 50, 65],
    'Department': ['QA', 'IT', 'Support'],
    'Salary': [50000, 60000, 70000]
}

dict_profile4 = {
    'Name': ['Alice4', 'Bob4', 'Charlie4'],
    'Age': [45, 50, 65],
    'Department': ['Designer', 'Marketing', 'IT'],
    'Salary': [50000, 60000, 70000]
}

df1 = pd.DataFrame(dict_profile1)
df2 = pd.DataFrame(dict_profile2)
df3 = pd.DataFrame(dict_profile3)
df4 = pd.DataFrame(dict_profile4)

dataframes_list_using_for_loop = []
dataframes_list_using_list = ['df1', 'df2', 'df3', 'df4']

for i in range(1, 5):
    dataframes_list_using_for_loop.append(globals()[f'df{i}'])

dataframes_object = []
# Using globals() to get the dataframe objects from the names in the list
for name in dataframes_list_using_list:
    dataframes_object.append(globals()[name])

# Concatenate dataframes
df1 = pd.concat(dataframes_list_using_for_loop, ignore_index=True)
print(f'concat dataframes_list_using_for_loop data: {df1} \n')

# Concatenate dataframes using list
df2 = pd.concat(dataframes_object, ignore_index=True)
print(f'concat dataframes_list_using_list data: {df2} \n')

# Show dataframe
df3 = pd.concat([df1, df2], ignore_index=True)
print(f'concat df1 , df2: {df3} \n')
```

#### checking types

```python
print(f'type of dataframes_list_using_for_loop {type(dataframes_list_using_for_loop)}')
print(f'type of df3 {type(df3)}')
```

```python
# appending data
New_entry = pd.DataFrame(
    [{
        'Name': 'David', 
        'Age': 28, 
        'Department': 
        'Marketing', 
        'Salary': 55000
    },
    {
        'Name': 'Eve', 
        'Age': 32, 
        'Department': 'Sales', 
        'Salary': 65000
    },
    {
        'Name': 'Frank', 
        'Age': 29, 
        'Department': 'IT', 
        'Salary': 72000
    }])
# Concat method instead of append method
df3 = pd.concat([df3, New_entry], ignore_index=True)
```
